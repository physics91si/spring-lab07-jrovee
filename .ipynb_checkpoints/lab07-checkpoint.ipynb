```python
from skimage import data, io
import numpy as np
from matplotlib import pyplot as plt
import time
```

## Part 1

```python
# Load image here. Display the image and the image shape.
starry = io.imread('starry.jpg')
print(starry.shape)
io.imshow(starry)
plt.show()
```

```python
# Make the color arrays and display them on a labelled subplot.
redCopy = np.copy(starry)
greenCopy = np.copy(starry)
blueCopy = np.copy(starry)

redCopy[:,:,1:3] = 0
greenCopy[:,:,0] = greenCopy[:,:,2] = 0
blueCopy[:,:,0:2] = 0

f, ((ax1,ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(10,10))
ax1.imshow(starry)
ax1.set_title('Original')
ax2.imshow(redCopy)
ax2.set_title('Red')
ax3.imshow(greenCopy)
ax3.set_title('Green')
ax4.imshow(blueCopy)
ax4.set_title('Blue')
plt.tight_layout()
plt.show()
```

## Part 2

```python
from math import sqrt
from skimage.feature import blob_log, blob_doh
from skimage.color import rgb2gray
```

```python
# Here is the image of the Hubble Deep Field. 
image = data.hubble_deep_field()[0:500, 0:500]

#Convert the image to grayscale and display next to the original.
grayImage = rgb2gray(image)
fig = plt.figure(figsize=(10,10))
fig.add_subplot(1,2,1)
io.imshow(image)
fig.add_subplot(1,2,2)
io.imshow(grayImage)
plt.show()
```

```python
import matplotlib
```

```python
# Apply the two Blob detection methods and measure the time taken by the algorithms.
gaussianStartTime = time.perf_counter()
log_data = blob_log(grayImage,min_sigma=1, max_sigma = 50)
gaussianEndTime = time.perf_counter()
hessianStartTime = time.perf_counter()
doh_data = blob_doh(grayImage, min_sigma = 1, max_sigma = 50)
hessianEndTime = time.perf_counter()

print("dog took " + str(gaussianEndTime-gaussianStartTime) + 
      " seconds and doh took " + str(hessianEndTime-hessianStartTime) + " seconds.")

```

```python
# Plot the 2 methods side-by-side along with circle patches. 
f, (ax1, ax2) = plt.subplots(1,2,figsize=(15,15))
ax1.imshow(grayImage)
for row in log_data:
    nextCircle = plt.Circle((row[1],row[0]), row[2]*sqrt(2),color = 'red',fill=False)
    ax1.add_patch(nextCircle)
ax2.imshow(grayImage)
for row in doh_data:
    nextCircle = plt.Circle((row[1],row[0]), row[2],color = 'red',fill=False)
    ax2.add_patch(nextCircle)
plt.show()
```
